Choose one of your favourite movie on IMDB website (e.g. for the movie "Top Gun: Maverick", the URL is :https://www.imdb.com/title/tt1745960/reviews?sort=totalVotes&dir=desc&ratingFilter=0

- `Function Input`: book page URL
- `Function Output`: save all reviews as a DataFrame of columns (`title, reviewer, date, rating, review, helpful`). E.g., for the given URL, you can get 25 reviews.

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.keys import Keys
import time
import re
```

```python
def getReviews(page_url):
    
    #fireFoxOptions = webdriver.FirefoxOptions()
    #fireFoxOptions.headless = True
    
    driver = webdriver.Firefox()     #options = fireFoxOptions,)
    driver.get(page_url)
    
    titles = driver.find_elements(By.CLASS_NAME,"title")
    users = driver.find_elements(By.CLASS_NAME,"display-name-link")
    dates = driver.find_elements(By.CLASS_NAME,"review-date")
    ratings = driver.find_elements(By.CLASS_NAME, "rating-other-user-rating")
    
    #expander = driver.find_elements(By.CLASS_NAME, "ipl-expander")
    #expander = reversed(expander)
    #expander = list(expander)[0:20]
#     for exp in expander:
#         print(exp
#         exp.location_once_scrolled_into_view
#         time.sleep(10)
#         #exp.click()
#         print("done")
#         time.sleep(10)
#         exp.click()

#    The location works by itself but when the click is there it doesn't work and there is an error
#    I thought it was because the box expanding changed the location of other blocks so I decided to go backwards
#    It still didn't work :(
#    After a lot of time I figured some blocks don't have expand button but it still tries to click it causing the error
#    Took a lot of time but eventually the try, except method worked


    expanders = driver.find_elements(By.CLASS_NAME, "ipl-expander")
    for exp in expanders:
        exp.location_once_scrolled_into_view
        try:
            exp.click()
        except:
            pass

        
    reviews = driver.find_elements(By.CLASS_NAME, "content") #"text show-more__control")
    
    review_text = []
    helpful = []

    for rev in reviews:
        temp = rev.text
        temp3 = temp.splitlines()
        temp4 = temp3[0:-2]
        review_text.append("".join(temp3[0:-2]))
        helpful.append(temp3[-2][:-42])
    
    
    def text(elem):
        temp = []
        for ele in elem:
            temp.append(ele.text)
        return temp
    
    titles_text = text(titles)
    users_text = text(users)
    dates_text = text(dates)
    
    ratings_text = []
    for rat in ratings:
        ratings_text.append(rat.text[:-3])
   

    titles_text = titles_text[:-1]
    reviews_tb = pd.DataFrame(
        {
        "title"   : titles_text,
        "user"    : users_text,
        "date"    : dates_text,
        "rating"  : ratings_text,
        "review"  : review_text,
        "helpful" : helpful
        }
    )
    driver.quit()
    return reviews_tb

```

```python
page_url = 'https://www.imdb.com/title/tt1745960/reviews?sort=totalVotes&dir=desc&ratingFilter=0'
#page_url = 'https://www.imdb.com/title/tt0468569/reviews?sort=totalVotes&dir=desc&ratingFilter=0'
getReviews(page_url)
```

```python
def getReviews2(page_url):
    

    driver = webdriver.Firefox()
    driver.get(page_url)

    for i in range(5):
        
        temp = driver.find_element(By.CLASS_NAME, "load-more-data")
        temp.location_once_scrolled_into_view
        temp.click()
        time.sleep(5)
    
    titles = driver.find_elements(By.CLASS_NAME,"title")
    users = driver.find_elements(By.CLASS_NAME,"display-name-link")
    dates = driver.find_elements(By.CLASS_NAME,"review-date")
    ratings = driver.find_elements(By.CLASS_NAME, "rating-other-user-rating")
    

    expanders = driver.find_elements(By.CLASS_NAME, "ipl-expander")
    for exp in expanders:
        exp.location_once_scrolled_into_view
        try:
            exp.click()
        except:
            pass

        
    reviews = driver.find_elements(By.CLASS_NAME, "content") #"text show-more__control")
    
    review_text = []
    helpful = []

    for rev in reviews:
        temp = rev.text
        temp3 = temp.splitlines()
        temp4 = temp3[0:-2]
        review_text.append("".join(temp3[0:-2]))
        helpful.append(temp3[-2][:-42])
    
    
    def text(elem):
        temp = []
        for ele in elem:
            temp.append(ele.text)
        return temp
    
    titles_text = text(titles)
    users_text = text(users)
    dates_text = text(dates)
    
    ratings_text = []
    count = 0
    for rat in ratings:
        ratings_text.append(rat.text[:-3])
        count += 1

    ratings_text = ratings_text[:-10] + [None] + ratings_text[-10:]
    titles_text = titles_text[:-1]
    
    reviews_tb = pd.DataFrame(
        {
        "title"   : titles_text,
        "user"    : users_text,
        "date"    : dates_text,
        "rating"  : ratings_text,
        "review"  : review_text,
        "helpful" : helpful
        }
    )
    driver.quit()
    return reviews_tb
    
```

```python
page_url = 'https://www.imdb.com/title/tt1745960/reviews?sort=totalVotes&dir=desc&ratingFilter=0'
review = getReviews2(page_url)
```
